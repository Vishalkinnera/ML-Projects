```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#LOAD THE DATASET

credit_card_data = pd.read_csv('/content/drive/MyDrive/Datasets/creditcard.csv')
credit_card_data.head()
```

```python
credit_card_data.tail()
```

WE have transactions of around 2 days

```python
#DATASET INFORMATION

credit_card_data.info()
```

```python
#CHECK the number of missing values in each column

credit_card_data.isna().sum()
```

```python
#Checking the distribution of Legit transaction and fraud transaction

credit_card_data['Class'].value_counts()
```

This data is very imbalanced.
There are 2 classes and more than 99% is in one class.
Machine cant recognise the fraud transcations.
We have to handle it

0----Normal Transaction
1----Fraud

```python
#Seperating the data for analysis

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#STATISTICAL MEASURES OF THE DATA

legit.Amount.describe()
```

```python
fraud['Amount'].describe()
```

```python
#compare the values for both transactions

credit_card_data.groupby('Class').mean()
```

```python
#UNDER-SAMPLING

#BUILD a sample dataset containing similar distribution of normal transaction and fraud
```

```python
#Number of Fraud transactions = 492
```

```python
legit_sample = legit.sample(n=492)
legit_sample
```

Concatenate 2 dataframes

```python
new_dataset = pd.concat([legit_sample, fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into X and y

```python
X = new_dataset.drop(columns="Class",axis=1)
y = new_dataset['Class']
```

```python
X.head()
```

```python
y
```

```python
#SPLITING DATA INTO TRAIN AND TEST
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING
Logistic Regression- used for Binary Classification Models

```python
model = LogisticRegression()
```

```python
#training the LogReg model with training data
model.fit(X_train,y_train)
```

MOdel Evaluation  based on Accuracy Score

```python
#Accuracy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy)
```

```python
#ACCURACY ON TEST DATA
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,y_test)
```

```python
print('Accuracy on test data:',test_data_accuracy)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(X_test_prediction,y_test)
f1
```
