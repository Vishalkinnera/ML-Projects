```python
#LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#IMPORT THE BOSTON HOUSEPRICE DATASET

house_price_dataset = sklearn.datasets.load_boston()
```

```python
print(house_price_dataset)
```

```python
#Loading the dataset to a pandas dataframe

house_price_dataframe = pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
#Add the target(Price) column to the dataframe

house_price_dataframe['price']=house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#Checking the shape of the dataframe

house_price_dataframe.shape
```

```python
house_price_dataframe.isna().sum()
```

```python
#STATISTICAL MEASURES OF  THE DATASET
house_price_dataframe.describe()
```

```python
#UNDERSTANDING THE CORRELATIO BETWEEN VARIOUS FEATURES IN THE DATASET
#2 TYPES OF CORRELATION
#1) POSITIVE----If one value increases, the other variables valuw also increases
#2) NEGATIVE
```

```python
correlation = house_price_dataframe.corr()
correlation
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(correlation,annot=True,fmt='.1f',cmap='Blues',linewidths=2)
```

```python
#SPLITTING THE DATA AND TARGET
```

```python
X = house_price_dataframe.drop(['price'],axis=1)
y = house_price_dataframe['price']
```

```python
X.head()
```

```python
y.head()
```

```python
#SPLIT INTO TRAIN AND TEST

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#MODEL TRAINING
#We use XGBOOST Regressor
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
#PREDICTION ON TRAINING DATA
#Accuracy for predicition on training data

training_data_predicition = model.predict(X_train)
```

```python
print(training_data_predicition)
```

```python
#R2 error for Regression

score_1 = metrics.r2_score(y_train,training_data_predicition)

#Mean Absolute Error

score_2 = metrics.mean_absolute_error(y_train,training_data_predicition)

print('R2 error :', score_1)
print('MAE:',score_2)
```

```python
#WE MADE PREDICTION ON TRAINING DATA
#NOW FOR TESTING DATA
```

```python
#PRediciton on Test data

test_data_predicition = model.predict(X_test)
```

```python
score_1 = metrics.r2_score(y_test,test_data_predicition)

#Mean Absolute Error

score_2 = metrics.mean_absolute_error(y_test,test_data_predicition)

print('R2 error :', score_1)
print('MAE:',score_2)
```

```python
#PLOT
#Visualise the Actual prices and Predicted prices

plt.scatter(y_train,training_data_predicition)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Price VS Predicted Price')
plt.show()
```

```python
plt.scatter(y_test,test_data_predicition)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Price VS Predicted Price')
plt.show()
```
